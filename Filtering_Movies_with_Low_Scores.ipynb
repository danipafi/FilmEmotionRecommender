```python
filtered_movies = clean_movies[(clean_movies['tomatoMeter'] < 30) & (clean_movies['audienceScore'] < 30)]
print(filtered_movies)
```

```python
grouped_movies = filtered_movies.groupby('id').agg({
    'tomatoMeter': 'mean',
    'audienceScore': 'mean',
    'title': 'first'  # MantÃ©m o tÃ­tulo do filme
}).reset_index()

grouped_movies
```

```python
# Count the total number of reviews per movie
total_reviews = clean_reviews.groupby('id').size()

# Count the total number of negative reviews per movie
negative_reviews = clean_reviews[clean_reviews['scoreSentiment'] == 'NEGATIVE'].groupby('id').size()

# Create a DataFrame with this information
sentiment_summary = pd.DataFrame({'total_reviews': total_reviews, 'negative_reviews': negative_reviews})

# Replace NaN values with 0 (in case some movies have no negative reviews)
sentiment_summary.fillna(0, inplace=True)

# Calculate the percentage of negative reviews
sentiment_summary['negative_percentage'] = (sentiment_summary['negative_reviews'] / sentiment_summary['total_reviews']) * 100

# Filter movies with more than 30% negative reviews
filtered_movies = sentiment_summary[sentiment_summary['negative_percentage'] > 30]

# Show the results
filtered_movies
```
