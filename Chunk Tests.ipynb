```python
import pandas as pd
pd.set_option('display.max_rows', None)  # Mostra todas as linhas

def has_common_genre(genres1, genres2):

    if 'Unknown' in genres1 or 'Unknown' in genres2:
        return True

    return not set(genres1).isdisjoint(genres2)


# Caminho para o arquivo do primeiro chunk
file_path = '/Users/danielebelmiro/Data_Analytics_Bootcamp/Rotten/processed_chunks/chunk_0.parquet'

# Carregar o chunk
chunk = pd.read_parquet(file_path)

# Verificar o número de linhas e colunas
print("\nNúmero de linhas e colunas:")
print(chunk.shape)

# Verificar os tipos de dados das colunas
print("\nTipos de dados das colunas:")
print(chunk.dtypes)

# Verificar se há correspondência de gêneros correta
print("\nVerificando correspondência de gêneros:")
for index, row in chunk.head().iterrows():
    print(f"id1: {row['id1']}, id2: {row['id2']}")
    print(f"genre: {row['genre']}, genre_2: {row['genre_2']}")
    print(f"Compartilham gêneros? {has_common_genre(row['genre'], row['genre_2'])}")
    print("-" * 50)
    
chunk
```
